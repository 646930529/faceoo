```python
import matplotlib.pyplot as plt
import numpy as np
a = np.random.rand(1,128)
b = np.random.rand(50,128)
```

```python
dst = np.linalg.norm(a-b, axis=1)
ang = np.dot(a,b.T)
plt.plot(dst,color='blue')
plt.plot(ang[0],color='red')
```

```python
def normalize(arr, axis=1, order=2):
    l2 = np.atleast_1d(np.linalg.norm(arr, order, axis))
    l2[l2 == 0] = 1
    return arr / np.expand_dims(l2, axis)
a = normalize(a)
b = normalize(b)
dst = np.linalg.norm(a-b, axis=1)
ang = np.dot(a,b.T)
plt.plot(dst,color='blue')
plt.plot(ang[0],color='red')
```
